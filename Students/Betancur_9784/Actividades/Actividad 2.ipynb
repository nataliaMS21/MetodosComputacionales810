# Problema 1: Persecucion policial

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xInicial = 0

vInicialMoto = float(input("inserte la velocidad de la moto: "))
aInicialPoli = float(input("inserte la aceleracion del policia: "))

t = np.linspace(0, 20, 500)
xMoto = xInicial + vInicialMoto*t
xPolicia = xInicial + (aInicialPoli * (t**2))/2
```

el punto de encuentro se da cuando xMoto = xPolicia, esto es

(vInicailMoto*2)/aInicialPoli

```python
prediccion = (vInicialMoto*2)/aInicialPoli
print(f'La prediccion de contacto con la velocidad inicial de la moto = {vInicialMoto} y la aceleracion inicial del policia = {aInicialPoli} es: {prediccion} ')

plt.axhline(0.0, color='black', alpha=0.5)
plt.axvline(0.0, color='black', alpha=0.5)

plt.plot(t, xMoto, label='Moto')
plt.plot(t, xPolicia, label='Policia')

plt.title("Policia vs Moto")
plt.xlabel("Tiempo s")
plt.ylabel("Distancia x")
plt.legend()
plt.grid(linestyle='-.', alpha=0.5)

plt.show()

print(f"En la grafica se confirma la prediccion, la posicion de la moto y el policia se cruzan en {prediccion}s")
```

# Problema 2:   Potencial LJ

#### a)

```python
def pLJ(E, r, = 1):

    Vr = 4*E * ((/r)**12 - (/r)**6)

    return Vr
```

```python
 = 1
E = 1
r = np.linspace(0.1,10, 10000)

Vr = pLJ(E, r, )

plt.axhline(0.0, color='black', alpha=0.5)
plt.axvline(0.0, color='black', alpha=0.5)

plt.plot(r, Vr, label='Potencial de Lennard-Jones')

plt.ylabel('Potencial')
plt.ylim(-1.1,1.1)

plt.xlabel('Distancia')
plt.xlim( -0.25,2.6*)

plt.legend()
plt.grid(linestyle='-.', alpha=0.5)
plt.show()
```

#### b y c)
La distancia de equilibrio sera el minimo de la curba de potencial, segun el problema deberia ser 2^(1/6)*, segun la grafica, se ve una diferencia minima    

```python
indMin = np.where(Vr == Vr.min())[0][0] #Encuentro el indice donde se encuentra el minimo de potencial
disMin = r[indMin]                      #Encuentro la distancia asociada a dico potendial

print(f'Con  = {} y E = {E}\n\nSegun la hipotesis rm = *2^(1/6) = {*2**(1/6)}.\nEn la grafica se puede ver que rm = {disMin}.\n\nSe diferencian en {abs(disMin-*2**(1/16))}')
```

# Problema 3: Datasets web

```python
import pandas as pd

# Dataset 1
# URL of the webpage
url = "https://worldpopulationreview.com/country-rankings/average-height-by-country"

# Read the HTML content
html_content1 = pd.read_html(url)

alturas = html_content1[0]

# Dataset 2
# URL of the webpage
url = "https://worldpopulationreview.com/country-rankings/average-net-worth-by-country"

# Read the HTML content
html_content2 = pd.read_html(url)

income = html_content2[0]
```

```python
#Traduccion y limpieza de alturas
alturasLimp = alturas.drop('Boys 19 (in)', axis= 1)
alturasLimp = alturasLimp.drop('Girls 19 (in)', axis= 1)
alturasLimp = alturasLimp.drop('Unnamed: 0', axis= 1)

alturasLimp = alturasLimp.rename(
    columns={
        'Country':'Pais',
        'Average Height, Boys, 19 yo (cm)':'Altura chicos (cm)',
        'Girls 19 (cm)':'Altura chicas (cm)'
})
```

```python
#traduccion y limpieza de income

incomeLimp = income.drop('Unnamed: 0', axis=1)
incomeLimp = incomeLimp.drop('Wealth per Adult 2022', axis=1)
incomeLimp = incomeLimp.drop('Total Wealth 2022', axis=1)
incomeLimp = incomeLimp.drop('Share of World Wealth 2022', axis=1)

incomeLimp = incomeLimp.rename(
    columns={
    'Country':'Pais',
    'GDP per Adult 2022':'PIB por adulto 2022'
})
```

```python
pd.options.plotting.backend = "matplotlib"

alturasLimpChicos = alturasLimp.sort_values('Altura chicos (cm)').drop('Altura chicas (cm)', axis=1)
alturasLimpChicos = alturasLimpChicos.drop(alturasLimpChicos[alturasLimpChicos['Pais'] == 'Total'].index[0], axis=0)

prim10Alturas = alturasLimpChicos.head(10)
ulti10alturas = alturasLimpChicos.tail(10)

Colombia = alturasLimpChicos[alturasLimpChicos['Pais'] == 'Colombia']
barras = ulti10alturas.merge(prim10Alturas, how='outer')
barras = pd.concat([barras, Colombia])
barras = barras.sort_values('Altura chicos (cm)')

barras.plot.bar(x= 'Pais', color= 'blue')
```
