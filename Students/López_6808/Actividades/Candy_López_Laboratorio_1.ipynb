Dise침ar un programa en el que entrado  un numero `a`  retorne una variable booleana True or false si cumple que es:

- [N칰meros defectivo](https://es.wikipedia.org/wiki/N칰mero_defectivo) : la suma de los divisores propios es menor que el n칰mero.

  La rutina se deber치 llamar n칰meros_defectivos



- [N칰meros abundantes](https://es.wikipedia.org/wiki/N칰mero_abundante): la suma de los divisores es mayor que el n칰mero.

  La rutina se deber치 llamar n칰meros_abundantes



- [N칰meros semiperfectos](https://es.wikipedia.org/wiki/N%C3%BAmero_semiperfecto) la suma de todos o algunos de los divisores propios es igual al n칰mero.

  La rutina se debera llamar numeros_semiperfectos

- [N칰meros perfectos](https://es.wikipedia.org/wiki/N%C3%BAmero_perfecto) la suma de todos sus divisores propios, excepto el mismo numero, es igual al n칰mero.

   La rutina se deber치 llamar numeros_perfectos


- [N칰meros primos](https://es.wikipedia.org/wiki/N%C3%BAmero_primo) el n칰mero es divisible unicamente por s칤 mismo y por 1.

   La rutina se deber치 llamar numeros_primos


**Problema no obligatorio**

Retornar los 30 primeros n칰meros de cada clase

````python
#Aqu칤 va su c칩digo
# Funci칩n para obtener los divisores propios de un n칰mero (excepto 칠l mismo)
def divisores_propios(n):
    return [i for i in range(1, n) if n % i == 0]

# 1. N칰meros defectivos
def numeros_defectivos(n):
    return sum(divisores_propios(n)) < n

# 2. N칰meros abundantes
def numeros_abundantes(n):
    return sum(divisores_propios(n)) > n

# 3. N칰meros semiperfectos
def numeros_semiperfectos(n):
    from itertools import combinations
    divisores = divisores_propios(n)
    for i in range(1, len(divisores) + 1):
        for combo in combinations(divisores, i):
            if sum(combo) == n:
                return True
    return False

# 4. N칰meros perfectos
def numeros_perfectos(n):
    return sum(divisores_propios(n)) == n

# 5. N칰meros primos
def numeros_primos(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Funci칩n para obtener los primeros 30 n칰meros de cada clase
def primeros_30(funcion_clasificadora):
    numeros = []
    n = 1
    while len(numeros) < 30:
        if funcion_clasificadora(n):
            numeros.append(n)
        n += 1
    return numeros
````

````python
print("Primeros 30 semiperfectos:", primeros_30(numeros_semiperfectos))
````

````python
print("Primeros 30 primos:", primeros_30(numeros_primos))
````

## Problema del escal칩n de potencial en Mec치nica Cu치ntica

Una part칤cula de masa $( m )$ se mueve en una dimensi칩n y encuentra un **escal칩n de potencial** como el que se muestra a continuaci칩n:



El potencial viene dado por:

$V(x) =
\begin{cases}
0, & x < 0 \\
V_0, & x \geq 0
\end{cases}$

Queremos analizar el comportamiento de la part칤cula cuando se enfrenta a este cambio repentino en el potencial, considerando dos casos distintos seg칰n su energ칤a $( E )$:

---

### 丘勇 Caso 1: Energ칤a mayor que el potencial $( E > V_0 )$

En este caso, la part칤cula tiene suficiente energ칤a para **cruzar el escal칩n**.

La funci칩n de onda general es:

- Para $( x < 0 )$:

$\psi_I(x) = Ae^{ik_1x} + Be^{-ik_1x}$

- Para $( x \geq 0 )$:

$\psi_{II}(x) = Ce^{ik_2x}$

donde:

$k_1 = \sqrt{\frac{2mE}{\hbar^2}}, \quad k_2 = \sqrt{\frac{2m(E - V_0)}{\hbar^2}}$

> El primer t칠rmino en $( \psi_I )$ es la onda incidente, el segundo la onda reflejada, y $( \psi_{II} )$ es la onda transmitida.

**Condiciones de frontera:**

En $( x = 0 )$, se imponen:

$\psi_I(0) = \psi_{II}(0), \quad \psi'_I(0) = \psi'_{II}(0)$

De ah칤 se obtienen las relaciones entre $( A, B, C )$ y los coeficientes de **reflexi칩n** y **transmisi칩n**:

$R = \left| \frac{B}{A} \right|^2, \quad T = \frac{k_2}{k_1} \left| \frac{C}{A} \right|^2$

---

### 丘勇 Caso 2: Energ칤a menor que el potencial $( E < V_0 )$

La part칤cula no tiene suficiente energ칤a para cruzar el escal칩n, pero en Mec치nica Cu치ntica **no se detiene**, sino que **penetra una cierta distancia** en la regi칩n prohibida.

- Para $( x < 0 )$:

$\psi_I(x) = Ae^{ikx} + Be^{-ikx}, \quad k = \sqrt{\frac{2mE}{\hbar^2}}$

- Para $( x \geq 0 )$:

$\psi_{II}(x) = Ce^{-\kappa x}, \quad \kappa = \sqrt{\frac{2m(V_0 - E)}{\hbar^2}}$

> El t칠rmino exponencial decreciente describe la **penetraci칩n evanescente** en la barrera.

**Condiciones de frontera** iguales a las del caso anterior permiten calcular los coeficientes.

En este caso:

- El coeficiente de **reflexi칩n** es $( R = 1 )$
- El de **transmisi칩n** $( T = 0 )$

pero **la funci칩n de onda no se anula inmediatamente** en $( x \geq 0 )$, lo que demuestra el fen칩meno de **tunelamiento cu치ntico**.

---

### 游 Conclusi칩n

El an치lisis del escal칩n de potencial muestra que incluso un cambio abrupto en el potencial no detiene totalmente una part칤cula cu치ntica. A diferencia de la mec치nica cl치sica, **la funci칩n de onda penetra en la regi칩n prohibida** aunque la probabilidad de encontrar la part칤cula all칤 decrezca exponencialmente.

Esto es una manifestaci칩n clara de la naturaleza ondulatoria de la materia y del **principio de incertidumbre** de Heisenberg.

````python
#Aqu칤 va su c칩digo
import numpy as np

# Constantes
hbar = 1.0545718e-34  # Constante de Planck reducida (J췅s)
m = 9.11e-31          # Masa del electr칩n (kg)

# Par치metros de entrada (puedes modificarlos o pedir al usuario que los ingrese)
E = 5.0  # Energ칤a de la part칤cula en eV
V0 = 4.0  # Altura de la barrera en eV
L = 1e-9  # Ancho de la barrera en metros

# Conversi칩n de eV a Joules
eV_to_J = 1.602e-19
E_J = E * eV_to_J
V0_J = V0 * eV_to_J

# C치lculo de T y R dependiendo de si E > V0 o E < V0
if E > V0:
    k1 = np.sqrt(2 * m * E_J) / hbar
    k2 = np.sqrt(2 * m * (E_J - V0_J)) / hbar
    T = (4 * k1 * k2) / (k1 + k2)**2
    R = 1 - T
else:
    k1 = np.sqrt(2 * m * E_J) / hbar
    kappa = np.sqrt(2 * m * (V0_J - E_J)) / hbar
    T = np.exp(-2 * kappa * L)
    R = 1 - T

# Impresi칩n de resultados
print(f"Energ칤a de la part칤cula: {E} eV")
print(f"Altura de la barrera: {V0} eV")
print(f"Ancho de la barrera: {L} m")
print(f"Probabilidad de transmisi칩n (T): {T:.5f}")
print(f"Probabilidad de reflexi칩n (R): {R:.5f}")
````

La 칩rbita en el espacio de un cuerpo alrededor de otro, como un planeta alrededor del Sol, no necesita ser circular. En general toma la forma de una elipse, con el cuerpo a veces m치s cerca y a veces m치s lejos. Si se conocen la distancia $\ell_{1}$ del acercamiento m치s pr칩ximo que hace un planeta al Sol, llamado perihelio, y su velocidad lineal $v_{1}$ en el perihelio, entonces cualquier otra propiedad de la 칩rbita puede calcularse a partir de estos dos datos de la siguiente manera.
La segunda ley de Kepler nos dice que la distancia $\ell_{2}$ y velocidad $v_{2}$ del planeta en su punto m치s distante, o afelio, satisfacen $\ell_{2}v_{2}=\ell_{1}v_{1}$. Al mismo tiempo, la energ칤a total, cin칠tica m치s gravitacional, de un planeta con velocidad $v$ y distancia $r$ del Sol est치 dada por
    $$
    E = \tfrac{1}{2}mv^{2} - G\frac{mM}{r},
    $$
    donde $m$ es la masa del planeta, $M=1.9891\times 10^{30}\,\mathrm{kg}$ es la masa del Sol, y $G=6.6738\times 10^{-11}\,\mathrm{m}^{3}\,\mathrm{kg}^{-1}\,\mathrm{s}^{-2}$ es la constante gravitacional de Newton. Dado que la energ칤a debe conservarse, se llega a que $v_{2}$ es la ra칤z m치s peque침a de la ecuaci칩n cuadr치tica
    $$
    v_{2}^{2} - \frac{2GM}{v_{1}\ell_{1}}v_{2} - \left[v_{1}^{2} - \frac{2GM}{\ell_{1}}\right] = 0.
    $$
    Una vez que se tiene $v_{2}$, podemos calcular $\ell_{2}$ usando la relaci칩n $\ell_{2}=\ell_{1}v_{1}/v_{2}$.

- Dados los valores de $v_{1}$, $\ell_{1}$, y $\ell_{2}$, otros par치metros de la 칩rbita vienen dados por f칩rmulas simples que pueden derivarse de las leyes de Kepler y del hecho de que la 칩rbita es una elipse:
$$
    \text{Semieje mayor:}  \quad a = \tfrac{1}{2}(\ell_{1}+\ell_{2}),
$$
$$
    \text{Semieje menor:}  \quad b = \sqrt{\ell_{1}\ell_{2}},
$$
$$
    \text{Periodo orbital:}  \quad T = \frac{2\pi ab}{\ell_{1}v_{1}},
$$
$$
    \text{Excentricidad orbital:}  \quad e = \frac{\ell_{2}-\ell_{1}}{\ell_{2}+\ell_{1}}.
$$

Escriba un programa que solicite al usuario ingresar la distancia al Sol y la velocidad en el perihelio, luego calcule e imprima las cantidades $\ell_{2}$, $v_{2}$, $T$, y $e$.

- Pruebe su programa calculando las propiedades de las 칩rbitas de la Tierra (para la cual $\ell_{1}=1.4710\times 10^{11}\,\mathrm{m}$ y $v_{1}=3.0287\times 10^{4}\,\mathrm{m}\,\mathrm{s}^{-1}$) y del cometa Halley ($\ell_{1}=8.7830\times 10^{10}\,\mathrm{m}$ y $v_{1}=5.4529\times 10^{4}\,\mathrm{m}\,\mathrm{s}^{-1}$). Entre otras cosas, deber칤a encontrar que el per칤odo orbital de la Tierra es un a침o y el del cometa Halley es alrededor de 76 a침os.


````python
#Aqu칤 va su c칩digo
import math

# Constantes
G = 6.6738e-11  # Constante de gravitaci칩n universal (m^3 kg^-1 s^-2)
M = 1.9891e30   # Masa del Sol (kg)

def calcular_orbita(l1, v1):
    # Coeficientes de la ecuaci칩n cuadr치tica
    A = 1
    B = -(2 * G * M) / (v1 * l1)
    C = -(v1**2 - (2 * G * M) / l1)

    # Resolver la ecuaci칩n cuadr치tica
    discriminante = B**2 - 4 * A * C
    if discriminante < 0:
        raise ValueError("No hay soluci칩n real para v2.")

    v2_pos = (-B + math.sqrt(discriminante)) / (2 * A)
    v2_neg = (-B - math.sqrt(discriminante)) / (2 * A)

    # Escoger la ra칤z positiva m치s peque침a
    v2 = min(filter(lambda x: x > 0, [v2_pos, v2_neg]))

    # Calcular l2
    l2 = l1 * v1 / v2

    # Semieje mayor y menor
    a = 0.5 * (l1 + l2)
    b = math.sqrt(l1 * l2)

    # Periodo orbital
    T = (2 * math.pi * a * b) / (l1 * v1)

    # Excentricidad
    e = (l2 - l1) / (l2 + l1)

    return l2, v2, T, e

# ENTRADA MANUAL

print("Ingrese los datos del objeto (en unidades SI):")
try:
    l1_usuario = float(input("Distancia al Sol en el perihelio (l1) en metros: "))
    v1_usuario = float(input("Velocidad en el perihelio (v1) en m/s: "))

    l2, v2, T, e = calcular_orbita(l1_usuario, v1_usuario)

    print(f"\n=== Resultados del objeto ingresado ===")
    print(f"Distancia en el afelio (l2): {l2:.3e} m")
    print(f"Velocidad en el afelio (v2): {v2:.3e} m/s")
    print(f"Periodo orbital (T): {T:.2f} s ({T/(60*60*24*365.25):.2f} a침os)")
    print(f"Excentricidad orbital (e): {e:.5f}")
except ValueError as err:
    print(f"Error: {err}")

# CASOS DE PRUEBA: TIERRA Y HALLEY

datos = {
    "Tierra": {
        "l1": 1.4710e11,
        "v1": 3.0287e4
    },
    "Cometa Halley": {
        "l1": 8.7830e10,
        "v1": 5.4529e4
    }
}

for nombre, valores in datos.items():
    l1 = valores["l1"]
    v1 = valores["v1"]
    l2, v2, T, e = calcular_orbita(l1, v1)

    print(f"\n=== {nombre} ===")
    print(f"Distancia en el afelio (l2): {l2:.3e} m")
    print(f"Velocidad en el afelio (v2): {v2:.3e} m/s")
    print(f"Periodo orbital (T): {T:.2f} s ({T/(60*60*24*365.25):.2f} a침os)")
    print(f"Excentricidad orbital (e): {e:.5f}")
````

$$

$$
$$
\text{Posici칩n vertical:} \quad y(t) = y_0 + v_{0y} t - \frac{1}{2}gt^2 = y_0 + v_0 \sin\theta \cdot t - \frac{1}{2}gt^2

````python
#Aqu칤 va su c칩digo

import numpy as np
import matplotlib.pyplot as plt

# Datos iniciales
v0 = 4.0  # velocidad inicial en m/s
theta_deg = 45  # 치ngulo en grados
theta_rad = np.radians(theta_deg)  # convertir a radianes
y0 = 0.0  # altura inicial en m
g = 9.81  # gravedad en m/s^2

# Componentes de la velocidad
v0x = v0 * np.cos(theta_rad)
v0y = v0 * np.sin(theta_rad)

# Tiempo total de vuelo: cuando y(t) = 0 (par치bola sim칠trica)
t_total = 2 * v0y / g

# Crear un arreglo de tiempo
t = np.linspace(0, t_total, num=100)

# Ecuaciones del movimiento
x = v0x * t
y = y0 + v0y * t - 0.5 * g * t**2

# Graficar la trayectoria
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Trayectoria del proyectil', color='blue')
plt.title('Movimiento parab칩lico de un proyectil')
plt.xlabel('Posici칩n horizontal x (m)')
plt.ylabel('Posici칩n vertical y (m)')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)  # l칤nea del suelo
plt.show()
````

Los polinomios de Legendre son de gran importancia en f칤sica, especialmente en la soluci칩n de la ecuaci칩n de Laplace en coordenadas esf칠ricas. Son fundamentales en la expansi칩n en arm칩nicos esf칠ricos, lo que los hace esenciales en problemas de mec치nica cu치ntica y electromagnetismo. Cada polinomio de Legendre $P_n(x)$ es un polinomio de grado $n$. Este puede ser expresado usando la F칩rmula de Rodrigues:
     
$$
P_n (x)=\frac{1}{n!\ 2^n} \frac{d^n}{dx^n} (x^2-1)^n, \ \ n \in  \mathbb{Z^{+}}+\{0\}
$$
        
- Utilizando la f칩rmula de Rodrigues halle $P_0(x)$ y $P_1(x)$.
-  Con la f칩rmula de recurrencia
$$
            P_{j+1}(x)=\frac{(2j+1)x P_j (x)-jP_{j-1}(x)}{j+1},
$$
elabore un c칩digo que permita calcular $P_n(x)$ para $n\geq2$.
- En una misma gr치fica muestre $P_n(x)$ para $1\leq n\leq5$.


## C치lculo de los Polinomios de Legendre

La f칩rmula de **Rodrigues** para los polinomios de Legendre es:

$$
P_n (x) = \frac{1}{n! \, 2^n} \frac{d^n}{dx^n} (x^2 - 1)^n
$$

### 1. C치lculo de $( P_0(x) )$ y $( P_1(x) )$ con la f칩rmula de Rodrigues

Primero, calculemos $(P_0(x))$ y $(P_1(x))$:

#### Para $(P_0(x))$:
Usando la f칩rmula de Rodrigues con $( n = 0 )$:

$$
P_0(x) = \frac{1}{0! \, 2^0} \frac{d^0}{dx^0} (x^2 - 1)^0 = 1
$$

Por lo tanto, $( P_0(x) = 1 )$.

#### Para $( P_1(x) )$:
Usando la f칩rmula de Rodrigues con $( n = 1 )$:

$$
P_1(x) = \frac{1}{1! \, 2^1} \frac{d}{dx} (x^2 - 1)^1 = \frac{1}{2} \cdot 2x = x
$$

Por lo tanto, $( P_1(x) = x )$.

### 2. C치lculo de $( P_n(x) )$ para $( n \geq 2 )$ usando la f칩rmula de recurrencia

La f칩rmula de recurrencia que nos dan es:

$$
P_{j+1}(x) = \frac{(2j+1)x P_j(x) - j P_{j-1}(x)}{j+1}
$$

Comenzamos con los valores base:
- $( P_0(x) = 1 )$
- $( P_1(x) = x )$

Ahora, podemos usar la f칩rmula de recurrencia para calcular $( P_2(x) )$, $( P_3(x) )$, $( P_4(x) )$ y $( P_5(x) )$.

````python
#Aqu칤 va su c칩digo
import numpy as np
import matplotlib.pyplot as plt

# Funci칩n para calcular P_n(x) usando la f칩rmula de recurrencia
def legendre_polynomial(n, x):
    if n == 0:
        return np.ones_like(x)  # P_0(x) = 1
    elif n == 1:
        return x  # P_1(x) = x

    # Inicializamos P_0(x) = 1 y P_1(x) = x
    P0 = np.ones_like(x)
    P1 = x

    # Usamos la f칩rmula de recurrencia para obtener P_n(x)
    for j in range(1, n):
        P_next = ((2 * j + 1) * x * P1 - j * P0) / (j + 1)
        P0 = P1
        P1 = P_next

    return P1

# Definir el rango de x
x = np.linspace(-1, 1, 400)

# Graficar P_n(x) para n=1 a n=5
plt.figure(figsize=(8, 6))

for n in range(1, 6):
    Pn = legendre_polynomial(n, x)
    plt.plot(x, Pn, label=f'P_{n}(x)')

# A침adir etiquetas y t칤tulo
plt.title('Polinomios de Legendre para n=1 a n=5')
plt.xlabel('x')
plt.ylabel('P_n(x)')
plt.legend()
plt.grid(True)

# Mostrar la gr치fica
plt.show()
````
